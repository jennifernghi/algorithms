Given the `head` of a sorted linked list, *delete all duplicates such that each element appears only once*. Return *the linked list **sorted** as well*.

**Example 1:**

![https://assets.leetcode.com/uploads/2021/01/04/list1.jpg](https://assets.leetcode.com/uploads/2021/01/04/list1.jpg)

```
Input: head = [1,1,2]
Output: [1,2]

```

**Example 2:**

![https://assets.leetcode.com/uploads/2021/01/04/list2.jpg](https://assets.leetcode.com/uploads/2021/01/04/list2.jpg)

```
Input: head = [1,1,2,3,3]
Output: [1,2,3]

```

**Constraints:**

- The number of nodes in the list is in the range `[0, 300]`.
- `100 <= Node.val <= 100`
- The list is guaranteed to be **sorted** in ascending order.

````python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def deleteDuplicates(self, head ) :
        if not head:
            return None
        
        curr = head
        prev = curr
        curr = curr.next
        
        while curr:
            if prev.val == curr.val:
                prev.next = curr.next
                curr = prev.next
            else:
                prev = curr
                curr = curr.next
        
        return head
    
 
def create_singly_inked_list(lst): 
    for i, v in enumerate(lst):
        lst[i] = ListNode(v)
        
    for i in range(0, len(lst)-1):
        lst[i].next = lst[i+1]
        
    return lst

def create_singly_inked_list_to_list(listnode):
    out = []
    
    for i in listnode:
        out.append(i.val)
    return out




a = create_singly_inked_list([0,1,2])
print()
````
